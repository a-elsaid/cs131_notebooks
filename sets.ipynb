# Sets
- Sets are a mathematical concept
- they group entitites together based on some common attribute
- Example: 
    - My Cars makers
    - My motorbikes makers

```python
C = {'Toyota', 'Honda', 'Suzuki', 'Ford', 'BMW', "BMW"}
M = {'Kawasaki', 'Honda', 'BMW'}
```

```python
print(C)
```

## Sets Operations
### Membership 
- 'Toyota' $\in$ C ==> True

```python
'Toyota' in C
```

## Subset
- {'Toyota', 'Ford'} $\subset$ C

```python
c = {'Toyota', 'Ford', 'Fiat'} 
c.issubset(C)
```

## Add
C + 'Cadillac' = {'Suzuki', 'Honda', 'Toyota', 'Ford', 'BMW', 'Cadillac'}

```python
C.add('Cadillac')
print(C)
```

- Sets do not have duplicates

```python
C.add('Toyota')
print(C)
```

## Union
- C $\cup$ M = {'Toyota', 'Honda', 'Suzuki', 'Ford', 'BMW', 'Kawasaki'}

```python
CM = C | M
print(CM)
```

## Intersection
- C $\cap$ M = {'Honda', 'BMW')

```python
C & M
```

## Difference
- C - M = {'Toyota', 'Suzuki', 'Ford', 'Cadillac'}

```python
C - M
```

## Symmetric Difference
- (C $\cup$ M) - (C $\cap$ M) = {'Toyota', 'Suzuki', 'Cadillac', 'Ford','Kawasaki'}

```python
C ^ M
```

## Remove

```python
C.remove('Ford')
print(C)
```

## Create Empty Set

```python
x = set()
print(x)
x.add('caterpiller')
print(x)
```

## Casting to Sets

```python
a_lst = [1,2,3,3,3]
a_set = set(a_lst)
print(a_set)
```

```python
b_tuple = (1,2,3,2)
print (b_tuple)
b_set = set(b_tuple)
print(b_set)
```

```python
c_str = "abcdc"
c_set = set(c_str)
print(c_set)
```

## Iterating over Sets

```python
for c in C:
    print(c)
```

```python
i = 0
while i<len(C):
    print(C[i])
    i+=1
```
