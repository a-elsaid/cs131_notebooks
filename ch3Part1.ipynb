## ``if`` Statement (Selection Control)

``if <test>``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement1>`` </br>

```python
if 4>3:
    print ("it is true that 4 is greater than 3")
```

## ``if`` ``else`` Statement 

``if <test>``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement1>`` </br>
``else``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement2>`` </br>

```python
a = 11
if a==10:
    print ("a=10")
else:
    print ("a!=10")
```

```python
if 3>4:
    print("3 is larger than 4")
else:
    print("3 is NOT larger than 4")
```

## Nested ``if`` Statement 

``if <test1>:`` </br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement1>`` </br>
``else``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ``if <test2>:``    </br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement2>`` </br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ``else:`` </br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement2>`` </br>


```python
x = -10
if x < 0:
    print(x, " is a negative number")
else:
    if x > 0:
        print(x, " is a positive number")
    else:
        print(x," is 0")
```

```python
score = 83
```

```python
if score >= 90:
    grade = 'A'
else:
    if score >= 80:
        grade = 'B'
    else:
        if score >= 70:
            grade = 'C'
        else:
            if score >= 60:
                grade = 'D'
            else:
                grade = 'F'
```

```python
print (grade)
```

## ``if`` ``elif`` Statement

``if <test1>``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement1>`` </br>
``elif <test2>``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement2>`` </br>
``else <test3>``: </br>
&nbsp;&nbsp;&nbsp;&nbsp;    ``<statement3>`` </br>

```python
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
```

```python
print (grade)
```
