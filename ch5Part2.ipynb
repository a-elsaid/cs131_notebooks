### `def` Statement

```
def <name> (arg1,arg2, ... argN):
   <statements>
   return <value>
```

- The `def` statement creates a function object and assigns it to a name.
- Python interpreter executes the function bodyeach time the function is called.
- The argument names will be assigned with the objects passed in the parenthesis at the point of a call.
- return ends function call and sends a result back to a caller

````python
def is_even(x):
    """
    Input: x, a positive int
    Returns True if x is even, otherwise False
    """
    print ("Priniting from inside is_even()")
    return x%2 == 0
````

````python
is_even(3)
````

### LEGB
- **LEGB** stands for Local, Enclosing, Global, Built-in. Python resolves scope in the order in which each scope appears in this list

- Local (L): The local, or current, scope. This could be the body of a function or the top-level scope of a script. It always represents the scope that the Python interpreter is currently working in.

- Enclosing (E): The enclosing scope. This is the scope one level up from the local scope. If the local scope is an inner function, the enclosing scope is the scope of the outer function. If the scope is a top-level function, the enclosing scope is the same as the global scope.

- Global (G): The global scope, which is the top-most scope in the script. This contains all of the names defined in the script that are not contained in a function body.

- Built-in (B): The built-in scope contains all of the names, such as keywords, that are built-in to Python. Functions such as round() and abs() are in the built-in scope. Anything that you can use without first defining yourself is contained in the built-in scope.

#### We can access a variable defined outside

````python
k = 3
def fun2():
    print(f"from INSIDE fun2(): k:{k}")
fun2()    
````

#### If a variable name is defined twice, one as a global variable, and another is local to a function, inside the function, the local one will be used.

````python
k = 3
def fun2():
    k = 5
    print(f"from INSIDE fun2(): k:{k}")
fun2()    
````

#### The code inside a function cannot modify a variable defined outside

````python
k = 3
def fun2():
    k+=2
    print(f"from INSIDE fun2(): k:{k}")
fun2()    
````

````python
k = 3
def fun2():
    global k
    k+=2
    print(f"from INSIDE fun2(): k:{k}")
fun2()    

print(f"from OUTSIDE fun2(): k:{k}")
````
