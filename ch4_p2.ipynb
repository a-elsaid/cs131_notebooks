```python
matrix[0]     # one row: first row
```

## Review:
### Immutable data-types (int, float, string, tuples)
- There is a lot going on when we type: `x = 3`
- First, an integer 3 is created and stored in memory
- A name x is created
- An reference to the memory location storing the 3 is then assigned to the name x
- So: When we say that the value of x is 3
- we mean that x now refers to the integer 3


```python
x = 3.0
x
```

```python
y = x
```

```python
x = 4
```

```python
y
```

#### What if you want to actually copy a list not to just reference the same list??


#### You can slice it:

```python
a = [1,2,3,4]
b = a[:]
a.append(55)
print(a)
print(b)
```

#### You can use the funciton `list()`

```python
b = list(a)
a.append(66)
print (a)
print(b)
```
