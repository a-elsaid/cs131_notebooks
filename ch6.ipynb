## Basic Types vs Object Types
- in other programming languages - like C++ and Java - integers, floats, characters, array, and boolean are basic types
- in python, all the types are objects. This provides functionalities to those types
- objects functionalities comes from attribute/member functions, called methods, that are part of the objects

### Example: `list` objects:
- Python `list` object has different functions like:
    - `count()`
    - `append()`
    - `insert(<index>, <value>)`
    - `remove(<value>)`
    - `reverse()`
    - **`sort()`**
- We can sort a list using the python function `sorted(<list>)`, which takes an arugment of type `list` and **returns a list**, or
- we can use the `list.sort()` function, which doesn't take an argument becaue it is part of the `list` object
- The 'period' or the 'dot operator' is used to access the member of the object
- `sort()` function in `list` is does a functionality against the `list` object; As if we are telling it <em>"Sort Yourself"</em>

```python
lst1 = [2,3,1,4,7,9,8,5,6]
lst2 = sorted(lst1)
print(lst1)
print(lst2)

lst1.sort()
print(lst1)
```

```python
class Address:
    def __init__(self, number, street, unit, city, zipcode):
        self.number = number
        self.street = street
        self.unit = unit
        self.city = city
        self.zipcode = zipcode
```

```python
class Person:
    def __init__(self, name, dob, address):
        self.name = name
        self.dof = dob
        self.address = address
```

```python
import random
class Employee(Person):
    def __init__(self, name:str, address:Address, dob:str, doe:str, marritial_status:str):
        super().__init__(name, dob, address)
        self.doe = doe
        self.marritial_status = marritial_status
        self.department = None        
        self.__real_evaluation = 0
        

    def department_description(self):
        return ("An Employee of this branch of the Our Company")
    
    
    def evaluation(self):
        """
        Function Employee.evaluation() returns a rigorious evaluation about the 
        employee performance taking many factors into consideration:
        years of service, number of served projects, peer review, manager review
        """
        
        return f"Mr./Ms. {self.name} Overall Evaluaition is {self.__rollAdice()}/6"
  


    ## Encapsulation
    def __rollAdice(self):
        return random.randint(1,6)
    
        
```

```python
## Enheritance
class DeprtA_emp(Employee):
    def __init__(self, name: str, address: Address, dof: str, doe: str, marritial_status: str):
        super().__init__(name, address, dof, doe, marritial_status)
        self.department = "A"
        
    ## Polymorphism
    def department_description(self):
        return f"{self.department} is my department... TAKE ME SERIOUSLLY"
```
