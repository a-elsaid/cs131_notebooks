## Data Representation

### 1- Numbers Decimal Representation

- Supports 0-9 for each digit 
- Each digit place is multiplied with 10 (number of supprted digits) to the power of its respective place

$ 5352 = 5 \times 10^3 + 3 \times 10^2 + 5 \times 10^1 + 2 \times 10^0$

$ 1011 = 1 \times 10^3 + 0 \times 10^2 + 1 \times 10^1 + 1 \times 10^0$

### 2- Numbers Octal Representation
- Supports 0-7 for each digit
- Each digit place is multiplied by 8 (number of suppported digits) to the power of its repective place

$ 5352 = 5 \times 8^3 + 3 \times 8^2 + 5 \times 8^1 + 2 \times 8^0 =>> 2,794$ 

$ 1011 = 1 \times 8^3 + 0 \times 8^2 + 1 \times 8^1 + 1 \times 8^0 = 521$

<span style="color:red; font-size:1.5em">Remember:  </span>
- $0\text{o}5352 \neq 5352$ </br>
But:
- $0\text{o}5352 = 2,794$ </br>
****
- $0\text{o}1011 \neq 1011$ </br>
But:
- $0\text{o}1011 = 521$

Let's Try:

```python
oct(2794)
```

```python
oct(521)
```

To transfer from Oct to Decimal:

```python
int(0o5352)
```

```python
int(0o1011)
```

### 3- Numbers Hexadecimal Representation: 

- Supports 0-9, a(10), b(11), c(12), d(13), e(14), f(15)
- Each digit place is multiplied by 16 (number of suppported digits) to the power of its repective place

$ 5352 = 5 \times 16^3 + 3 \times 16^2 + 5 \times 16^1 + 2 \times 16^0 = 21,330$

$ 1011 = 1 \times 16^3 + 0 \times 16^2 + 1 \times 16^1 + 1 \times 16^0 = 4,113$

$\text{ad03} = 10 \times 16^3 + 13 * 16^2 + 0 \times 16^1 + 3 \times 16^0 = 4,4291$

<span style="color:red; font-size:1.5em">Remember:  </span>
- $0\text{x}5352 \neq 5352$ </br>
But:
- $0\text{x}5352 = 21,330$ </br>
****
- $0\text{x}1011 \neq 1011$ </br>
But:
- $0\text{x}1011 = 4,113$
****
- $0\text{xad03} = 4,4291$ </br>
