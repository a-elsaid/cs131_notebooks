### Example: Frequency of Words

```python
import re
def frequency(s):
    """Given a string of words, returns a dictionary of word frequencies"""
    freqDict = {} # initialize accumulator as empty dict
    for word in s.lower().replace("\n", " ").split(" "):
        # word = re.sub(r'[^a-zA-Z0-9]', '', word)
        
        # check whether word not in dictionary
        if word not in freqDict:
            freqDict[word] = 1
        else:
            freqDict[word] += 1
            
    return freqDict
```

## Properties of Python Dictionaries 

### 1- Keys are Unique 

```python
dict1 = {"a": 1, "b": 2, "c": 3}
dict1
```

```python
dict1 = {"a": 1, "b": 1 , "c": 3}
dict1
```

```python
dict1 = {"a": 1, "a": 2, "c":3}
dict1
```

### 2- Values are Not Unique

```python
dict2 = {"a": 2, "b": 2, "c":3}

print (f"dic1:{dict2}")

print (f"Value of key 'a' in dic1 is the same as value in key 'b': {dict2['b'] is dict2['b']}")
print (id(dict2['a']))
print (id(dict2['b']))
```

```python
dict2 = {"a": [2], "b": [2], "c":3}

print (f"dic1:{dict2}")

print (f"Value of key 'a' in dic1 is the same as value in key 'b': {dict2['b'] is dict2['b']}")
print (id(dict2['a']))
print (id(dict2['b']))
```

### 3- Order of entry is ~~not~~ preserved 

```python
dict3 = {'b':2, 'a':1, 'c':3}
dict3
```

### 4- Can not be sliced 

```python
dict3[1:]
```

### 5- Keys and Values can be Heterogeneous 

```python
dict4 = {1:"One", "Two": 2, "lst": [1,2,3,4], (1,2):"a tuple", "another tuple": (11,22)}
dict4
```

### 6- Dictionary are mutable 

```python
cars = {1:"Ford", 2:"Mercedes", 3:"Tesla", 4:"Jaguar", 5:"Nissan"}
print (id(cars))
cars[6] = "Lada"
print (id(cars))
```

### Why Use Dictionaries?

```python
import time
someLetters = ['a', 'l', 'q', 's', 'y', 'z']*1000000  # List with 6M repeated letters
print("Number of queries", len(someLetters))
```

```python
lettersScore
```

```python
# generate list of letters and scores
letters = list(lettersScore.keys())
scores = list(lettersScore.values())

# time using list operations to compute total score
startTime = time.time()
totalScore = 0

for query in someLetters:
    index = letters.index(query)
    totalScore += scores[index]

endTime = time.time()
timeList = endTime - startTime
print("Time taken using a list", round(timeList, 3), "seconds")
```

```python
# time using dictionaries to compute total score
startTime = time.time()
totalScore = 0

for letter in someLetters: 
    totalScore += lettersScore.get(letter, 1)

endTime = time.time()
timeDict = endTime - startTime
print("Time taken using a dictionary", round(timeDict, 3), "seconds")
```
