### Lists are:
- most flexible ordered collection of arbitrary objects (numbers, strings, lists)
    - a place to collect other objects
    - ordered left- to-right 
- Accessed by offset
    - indexing, slicing and concatenation 
- Variable length, heterogeneous, arbitrarily nestable
    - unlike strings, list can grow and shrink in place (!!)
    - lists may contain any type of object
    - you can create lists of lists

#### Strings are treated by Python like a sequence of charaters (kind of like list BUT NOT LISTS)

```python
string1 = "This is a one line string for testing"
```

```python
string1[4:30]
```

```python
string1[4:25:4]
```

#### However, you can cast a string to a list of characters:

```python
list_string1 = list(string1)
list_string1
```

```python
list_string1[4:25:4]
```

####  You can then do list operations on the string converted to list

```python
list1 = L0 + string1[2:5]
```

```python
list1 = L0 + list_string1[2:5]
list1
```
